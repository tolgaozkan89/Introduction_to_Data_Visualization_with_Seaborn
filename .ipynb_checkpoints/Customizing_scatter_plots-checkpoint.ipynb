# Customizing Scatter Plot

##  ----  Quick Summary ----

## Scatter plot overview
### Show relationship between two quantitative variables
---
### We've seen:
### 1. Subplots ( col and row )
### 2. Subgroups with color ( hue ) 
---
### New Customizations:
### 1. Subgroups with point size and style
### 2. Changing point transparency
### Use with both scatterplot() and relplot()
### ci = "sd" , ci = None

```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
```

# PART1: "tips" DATASET

```python
tips = sns.load_dataset("tips")
```

## 1.1 Subgroups with point size (size)

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",size="size")
plt.show()
```

# 1.2 Point size and hue (hue)

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",size="size",hue="size")
plt.show()
```

# 1.3 Subgroups with point style (style="smoker")

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",hue="smoker",style="smoker")
plt.show()
```

# 1.4 Changing point transparency

```python
sns.relplot(x="total_bill",y="tip",data=tips,kind="scatter",alpha=0.4)
plt.show()
```

# PART2: "mpg" DATASET

In this exercise, we'll explore Seaborn's mpg dataset, which contains one row per car model and includes information such as the year the car was made, the number of miles per gallon ("M.P.G.") it achieves, the power of its engine (measured in "horsepower"), and its country of origin.

What is the relationship between the power of a car's engine ("horsepower") and its fuel efficiency ("mpg")? And how does this relationship vary by the number of cylinders ("cylinders") the car has?

### Let's continue to use relplot() instead of scatterplot() since it offers more flexibility.

```python
mpg = pd.read_csv("mpg.csv")
```

```python
mpg.head()
```

## Changing the size of scatter plot points

Use relplot() and the mpg DataFrame to create a scatter plot with "horsepower" on the x-axis and "mpg" on the y-axis. Vary the size of the points by the number of cylinders in the car ("cylinders").

## size

```python
sns.relplot(x="horsepower",y="mpg",data=mpg,kind="scatter",size="cylinders")
plt.show()
```

To make this plot easier to read, use hue to vary the color of the points by the number of cylinders in the car ("cylinders").

## hue

```python
sns.relplot(x="horsepower",y="mpg",data=mpg,kind="scatter",size="cylinders",hue="cylinders")
plt.show()
```

## kind = line

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line")
plt.show()
```

## standard deviation

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line", ci="sd")
plt.show()
```

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line", ci=None)
plt.show()
```
