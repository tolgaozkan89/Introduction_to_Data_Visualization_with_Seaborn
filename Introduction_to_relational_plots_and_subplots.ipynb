```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
```

# PART1: "student-alcohol-consumption" DATASET

## Q: Why use relplot() instead of scatterplot() ? / 
## A: relplot() lets you create subplots in a single 

```python
student_data = pd.read_csv("student-alcohol-consumption.csv")
```

```python
student_data.head()
```

```python
student_data.columns
```

# 1. Creating subplots with col and row

# 1.1 scatterplot

```python
sns.scatterplot(x="absences", y="G3",data=student_data)
plt.show()
```

# 1.2 relplot

Modify the code to use relplot() instead of scatterplot().

```python
sns.relplot(x="absences", y="G3",data=student_data,kind="scatter")
plt.show()
```

# 1.2 + col="study_time"

Modify the code to create one scatter plot for each level of the variable "study_time", arranged in columns.

```python
sns.relplot(x="absences",y="G3",data=student_data,kind="scatter",col="study_time")
plt.show()
```

# 1.3 + row="study_time"

Adapt your code to create one scatter plot for each level of a student's weekly study time, this time arranged in rows.

```python
sns.relplot(x="absences",y="G3",data=student_data,kind="scatter",row="study_time")
plt.show()
```

# 2. Creating two-factor subplots

Use relplot() to create a scatter plot with "G1" on the x-axis and "G3" on the y-axis, using the student_data DataFrame.

```python
sns.relplot(x="G1",y="G3",data=student_data,kind="scatter")
```

Create column subplots based on whether the student received support from the school ("schoolsup"), ordered so that "yes" comes before "no".

```python
sns.relplot(x="G1",y="G3",data=student_data, kind="scatter",col="schoolsup",col_order=["yes","no"])
plt.show()
```

Add row subplots based on whether the student received support from the family ("famsup"), ordered so that "yes" comes before "no". This will result in subplots based on two factors.

```python
sns.relplot(x="G1",y="G3",data=student_data, kind="scatter",col="schoolsup",col_order=["yes","no"],row="famsup",row_order=["yes","no"])
plt.show()
```
